```python
#padansとnumpyをダウンロード
import os
import pandas as pd
import numpy as np
```

```python
path = os.path.join(path, 'Reviews.csv')
path
```

```python
df = pd.read_csv(path)
df
```

```python
#ProductIdごとにデータをまとめて、件数を数え、その中で一番多い件数を取り出す
df.groupby('ProductId').count().max()
```

```python
#最もコメント（行数）が多い商品（ProductId）と、そのコメント数を表示する

#ProductId が同じ行をまとめ、コメント数が最も多いProductId を取得
comment_counts = df.groupby('ProductId').size()
most_commented_product_id = comment_counts.idxmax()
max_comments = comment_counts.max()

print(f"Product ID with the most comments: {most_commented_product_id}")
print(f"Number of comments: {max_comments}")
```

```python
#一番多い件数を取り出す
df[df['ProductId'] == most_commented_product_id]
```

```python
#最もレビューが多い商品の「レビュー本文（Text）」と「評価点数（Score）」をそれぞれ抽出
#textのカラムを抽出
target = df[df['ProductId'] == most_commented_product_id]['Text'].tolist()

#scoreのカラムを抽出
df[df['ProductId'] == most_commented_product_id]['Score']
```

```python
#日本語で感情分析を行うために以下のライブラリをインストール
!pip install fugashi ipadic
!pip install unidic-lite
```

```python
#BERTで感情分析
from transformers import pipeline

classifier = pipeline(
    task="sentiment-analysis",
    model="koheiduck/bert-japanese-finetuned-sentiment",
    tokenizer="koheiduck/bert-japanese-finetuned-sentiment"
)
```

```python
sentences = target
results = classifier(sentences, truncation=True)  #トークン制限がある為、レビューの内、一部しか活用できない

for sentence, result in zip(sentences, results):
    print(f"sentence: {sentence}")
    print(f"label: {result['label']}, score: {round(result['score'], 4)}")
    print("-" * 30)
```

```python
positive_reviews = []

#[i]['label']はi番目の結果の中の "label" を取り出す
for i in range(len(results)):
    if results[i]['label'] == 'POSITIVE':
        positive_reviews.append(target[i])

print("positive Reviews:")
for i in positive_reviews:
    print(i)
```

```python
import re
import MeCab

mecab_tagger = MeCab.Tagger()

#
sentences = ",".join(positive_reviews)
node = mecab_tagger.parseToNode(sentences)
vocab_list = []

while node:
    word = node.surface
    hinshi = node.feature.split(",")[0]
    if hinshi == "名詞":
        if (not word.isnumeric()) and (not re.match(r'^[\u3040-\u309F]+$', word)):
            # 名詞が数値と平仮名のみの場合は除き、それ以外の名詞を保存
            vocab_list.append(word)
    else:
        pass
    node = node.next

print(vocab_list)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# tfidfモデルの作成と学習
tfidf_model = TfidfVectorizer(token_pattern='(?u)\\b\\w+\\b', norm=None)
tfidf_model.fit(vocab_list)
```

```python
# 対象のテキストをtf-idf値に変換
vocab_text = " ".join(vocab_list)
tfidf_vec = tfidf_model.transform([vocab_text]).toarray()[0]

# 単語: tf-idf値となるdictに変換
tfidf_dict = dict(zip(tfidf_model.get_feature_names_out(), tfidf_vec))

# tf-idf値が正のみの単語を残す
tfidf_dict = {word: num_val for word, num_val in tfidf_dict.items() if num_val > 0}

tfidf_dict
```
