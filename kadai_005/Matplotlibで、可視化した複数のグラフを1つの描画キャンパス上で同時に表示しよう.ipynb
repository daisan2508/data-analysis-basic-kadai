```python
import matplotlib.pyplot as plt
import numpy as np

# プロット
fig, axes = plt.subplots(2,3,tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

axes[0,1].plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# モーターから観測した音の周波数
hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]
# 箱ひげ図
axes[0,2].boxplot(hertz)

# データを生成する
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

axes[1,0].scatter(x, y)

# 画像を表示する
plt.show()
```
