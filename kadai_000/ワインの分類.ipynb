```python
#Pandasをインポート
import pandas as pd

#データの取得
from sklearn.datasets import load_wine

#変数に代入
dataset = load_wine()
dataset
```

```python
#データの値を確認
df = load_wine(as_frame=True).frame
df.head()
```

```python
#外れ値の確認
df.hist(figsize=(20, 15), bins=30)
```

```python
#モデル学習にどの説明変数を使用するかも含めて検討
#目的変数Yはtarget
#説明変数Xはalcohol
#学習データとテストデータに分割
X = df[['alcohol', 'od280/od315_of_diluted_wines']].to_numpy()
y = df['target'].to_numpy()

#train_test_split関数をインポートし、168と10に学習データとテストデータに分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10, random_state=0)

#分割後のデータ数
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#精度が良いモデルを最終モデルとして使用
#他の分類モデルがあれば、自分で調べて追加
#決定木とランダムフォレストはいくつかのハイパーパラメータ（木の深さ、葉の数など）を試す


#予測モデルの学習

#不明点：テキストにある「実行ごとに結果が変わることを避けたい場合は、random_stateを指定」の意味
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# 比較するモデル一覧
models = {
    "DecisionTreeClassifier": DecisionTreeClassifier(random_state=0),
    "RandomForestClassifier": RandomForestClassifier(random_state=0),
    "LogisticRegression": LogisticRegression(random_state=0)
}

# 最良モデルを探索
best_model = None #モデルインスタンス
best_model_name = None #モデル名
best_score = float('-inf') #スコア

for name, model in models.items():
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    print(f"{name}: R2 = {score:.2f}")

    # 現在のベストスコアを上回れば、最良モデルを更新する
    if score > best_score:
        best_score = score
        best_model_name = name
        best_model = model

print(f"\n最も精度が良いモデルは: {best_model_name}, スコア: {best_score}")
```

```python
#決定係数の表示
print(best_model.score(X_train, y_train))
print(best_model.score(X_test, y_test))
```
