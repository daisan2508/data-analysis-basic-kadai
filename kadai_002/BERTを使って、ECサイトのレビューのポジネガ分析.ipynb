```python
negative_reviews = []

#len()é–¢æ•°ã®æ„å‘³ã€‚range(len(results))ã§ï¼‘å€‹ãšã¤åˆ¤å®šã—ã¦ã„ãï¼Ÿ
#[i]['label']ã¯iç•ªç›®ã®çµæžœã®ä¸­ã® "label" ã‚’å–ã‚Šå‡ºã™
for i in range(len(results)):
    if results[i]['label'] == 'NEGATIVE':
        negative_reviews.append(sentences[i])

print("Negative Reviews:")
for i in negative_reviews:
    print(i)
```

```python
import re
import MeCab

mecab_tagger = MeCab.Tagger()

#
sentences = ",".join(negative_reviews)
node = mecab_tagger.parseToNode(sentences)
vocab_list = []

while node:
    word = node.surface
    hinshi = node.feature.split(",")[0]
    if hinshi == "åè©ž":
        if (not word.isnumeric()) and (not re.match(r'^[\u3040-\u309F]+$', word)):
            # åè©žãŒæ•°å€¤ã¨å¹³ä»®åã®ã¿ã®å ´åˆã¯é™¤ãã€ãã‚Œä»¥å¤–ã®åè©žã‚’ä¿å­˜
            vocab_list.append(word)
    else:
        pass
    node = node.next

print(vocab_list)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

# tfidfãƒ¢ãƒ‡ãƒ«ã®ä½œæˆã¨å­¦ç¿’
tfidf_model = TfidfVectorizer(token_pattern='(?u)\\b\\w+\\b', norm=None)
tfidf_model.fit(vocab_list)
```

```python
# å¯¾è±¡ã®ãƒ†ã‚­ã‚¹ãƒˆã‚’tf-idfå€¤ã«å¤‰æ›
vocab_text = " ".join(vocab_list)
tfidf_vec = tfidf_model.transform([vocab_text]).toarray()[0]

# å˜èªž: tf-idfå€¤ã¨ãªã‚‹dictã«å¤‰æ›
tfidf_dict = dict(zip(tfidf_model.get_feature_names_out(), tfidf_vec))

# tf-idfå€¤ãŒæ­£ã®ã¿ã®å˜èªžã‚’æ®‹ã™
tfidf_dict = {word: num_val for word, num_val in tfidf_dict.items() if num_val > 0}

tfidf_dict
```
